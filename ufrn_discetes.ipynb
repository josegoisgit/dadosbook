## Class

```python
class Dataship( dict ):
    '''
        https://www.pirateglossary.com/
        https://translate.google.com/translate?sl=en&tl=pt&u=https://www.pirateglossary.com/
    '''
    
    sailors = dict()
    descriptor = ''
    
    def __init__(self):
        self = super().__init__( self )

    def __init__(self, relatives ):
        self.relatives = relatives

    def envision_crew ( self, crew ):
        if not crew in self.sailors:
            self.sailors[crew]            = dict()
            self.sailors[crew]['members'] = set()
            self.sailors[crew]['files']   = set()    

    def encrew_visitor( self, crew, visitor ):
        self.envision_crew( crew )
        self.sailors[crew]['members'].add(visitor)
        print('\tmember ' + visitor,end='\n\n')
        for file in self.relatives[visitor]:
            if is_included( file ):
                print('\t\tfile '+file)
                self.sailors[crew]['files'].add( file )
        print('')

    def listen(self, crew, meeting, is_included):
        print( 'crew ' + crew, sep='', end='\n\n' )
        for visitor in meeting:    
            if visitor in self.relatives:
                self.encrew_visitor ( crew, visitor )                
                
        
    def listen_expression(self, crew, calls, is_included):
        print( 'crew ' + crew, sep='', end='\n\n' )
        for call in calls:
            visitors = [v for v in self.relatives.keys() if re.findall(call,v)]
            for visitor in visitors:
                self.encrew_visitor ( crew, visitor )                    

    def load( self ):
        import pandas as pd 
        d = lambda f: { 'filepath_or_buffer':f, 'sep':';', 'decimal': ',' }
        print('Crew', sep='', end='\n\n')
        for crew, sailors in self.sailors.items():
            print( crew , sep='', end='\n\n')
            
            data = []
            
            print('\t', 'Files', sep='', end='\n\n')
            for file in sailors['files']:    
                print('\t', file, sep='', end=' ')
                t = 0
                while True:
                    t += 1
                    print(t,sep='',end=' ')
                    try:
                        data.append( pd.read_csv( **d( file) ) )
                        break
                    except:
                        continue
                print()
            print('',end='\n\n')
            self[crew] = pd.concat( data ).reset_index(drop=True)

    def ahoy( self, crew ):
        if crew in globals():
            self[ crew ] = globals()[ crew ]
    
    def parley( self ):
        for member in self.keys():
            globals()[ member ] = self[ member ]

    def ay( self, crew ):
        if crew in globals() and crew in self.keys():
            self[ crew ] = globals()[ crew ]

    def heave( self ):
        for member in self.keys():
            if member in globals():
                self[ member ] = globals()[ member ]

    def belay( self ):
        for member in self.keys():
            if member in globals():
                del globals()[ member ]

    def avast( self, member ):
        super( variable_dict, self ).pop( member )
        if member in globals():
            del globals()[ member ]

    def ho( self, crew, member, function, kwarg_function ): 
        return self[ crew ][ member ].apply( lambda entry: function( kwarg_function( entry ) ) )  

    def chase( self, crew, member, function, kwarg_function ): 
        id = self[ crew ][ member ].apply( lambda entry: function( kwarg_function( entry ) ) )  
        return self[ crew ][ id ]

    def crimp( self, crew, member, function, kwarg_function ): 
        self[ crew ][ member ] = self[ crew ][ member ].apply( lambda entry: function( kwarg_function( entry ) ) )  
```

```python
def is_included( filename ):
    get_ap_str  = lambda filename: re.sub("[^\d']",'',os.path.basename(filename))

    get_a       = lambda filename: int(get_ap_str(filename)[:4])
    get_p       = lambda filename: int(get_ap_str(filename)[-1])
    get_len     = lambda filename: len( get_ap_str(filename) )

    is_ano      = lambda filename: get_a(filename) in anos_analise
    is_periodo  = lambda filename: get_p(filename) in periodos_analise

    is_within   = lambda filename: is_ano(filename) & is_periodo(filename)

    v = filename.endswith('.csv')

    if get_len(filename) == 4:
        return is_ano(filename) & v

    if get_len(filename) == 5:
        return is_ano(filename) & is_periodo(filename) & v
    
    return v
```

#### figure_histogram

```python
def figure_histogram(varname, key,_fdict={}):
    print('fig')
    if ( isinstance(key,str) or ( isinstance(key,list) and ( len(key) == 1 ) ) ):
        key_counts = dataship[varname][key].value_counts()
        _title = ''.join(key) + ' de ' + varname
    
    if ( isinstance(key,list) and ( len(key) > 1 ) ):
        key_counts = dataship[varname].groupby(key[:-1])[key[-1]].value_counts()
        __title = key[-1] + ' de ' + varname
        _title = __title

        if isinstance(key,str):
            _title = _title + key    
        
        if isinstance(key,list):
            if len(key) > 1:
                _title = _title + ' agrupado por '

            if len(key) == 2:
                _title = _title + key[0]

            if len(key) == 3:
                _title = _title + key[0] + ' & ' + key[1]

            if len(key) >= 4:
                _title = _title + ', '.join(key[:-2]) + ' & ' + key[-2]

    y,key_count = zip(*enumerate(key_counts.items()))
    label, count = zip(*key_count)
    label = [str(l) for l in label]
    f = lambda x,n: np.ceil(x/(10**n))*(10**n)
    ymax = f(count[0],3)
    #ymax = 10**(np.ceil(np.log10(count[0])*100)/100)
    f_dict = {
        'title' : _title,
        'height' : 400,
        'width' : 400,
        'y_minor_ticks' : 2,
        'x_axis_label' : varname + ' entries',
        'x_range' :(0,ymax),
        'y_range' :label
    }
    f_dict.update(_fdict)

    return label,count,ymax,f_dict
```

#### horizontal_histogram

```python
def horizontal_histogram(varname,key,figure_dict_overwrite={},glyph_dict_overwrite={},inc_range=range(10), do_show=True, do_order=True):
    
    label, count, ymax, figure_dict = figure_histogram(varname, key, figure_dict_overwrite)
    label,count = zip(*[(l,c) for (e,l,c) in zip(list(range(len(label))),label,count) if e in inc_range])

    if do_order:
        label,count = zip(*sorted(zip(label,count),key=lambda z:z[1]))

    figure_dict.update(figure_dict_overwrite)

    glyph_dict = {
        'y' : label,
        'height' : .5,
        'right' : count,
        'fill_color' : 'black',
        'fill_alpha' : 1,
    }
    glyph_dict.update(glyph_dict_overwrite)

    f = figure(**figure_dict)
    
    b = getattr(f,'hbar')(**glyph_dict)
    
    if do_show: show(f)

    return label,count,f,b
```

#### vertical_histogram

```python
def vertical_histogram(varname,key,_fdict={},_gdict={},inc_range=range(10),do_order=True):
    __fdict = {'y_axis_label' : varname + ' entries'}
    _fdict.update(__fdict)

    label, count, ymax, f_dict = figure_histogram(varname, key, _fdict)        
    label,count = zip(*[(l,c) for (i,l,c) in zip(lrange(label),label,count) if i in inc_range])
    if do_order:
        label,count = zip(*sorted(zip(label,count),key=lambda z:z[1]))

    f_dict.update({'x_range':label, 'y_range' :(0,ymax)})
    f_dict.update(_fdict)

    vbar_dict = {
        'x' : label,
        'width' : .5,
        'top':count,
        'fill_color' : 'black',
        'fill_alpha' : 1,
    }
    vbar_dict.update(_gdict)

    f = figure(**f_dict)

    f.vbar(**vbar_dict)
    
    show(f)
    return label,count,f
```

### Funções

#### HTML

```python
def get_html(url):
    from lxml.etree import HTML
    import requests

    requested = requests.request('GET', url)
    html      = HTML(requested.text)    

    return html
```

#### XPath

```python
def contains_clause(dic, operator='OR'):
    predicados = ["contains(@{k},'{v}')".format(k=k,v=v) for k,v in dic.items()]

    return operator.join(predicados)
```

```python
def compose_xpath(element,attribute,content):
    dictionary = contains_clause({attribute:content})

    str_dict = {
        'element' : element,
        'dictionary' : dictionary 
    }

    return "{element}[{dictionary}]".format(**str_dict)
```

```python
def get_xnodes(html, element, attribute='class', content='', complement=''):
    if complement:
        xnode = html.xpath(compose_xpath(element,attribute,content) + complement)
    else:
        xnode = html.xpath(compose_xpath(element,attribute,content) )
    return xnode if xnode else None
```

```python
def get_xnode(html, element, attribute='class', content='', complement=''):
    xnodes = get_xnodes(html, element, attribute, content, complement)
    return xnodes[0] if xnodes else None
```

#### Query

#### Page counts

```python
def get_html_page_count(html):
    xnode = get_xnode( html, '//div', 'class', 'pagination' ) 
    if not xnode:
        return 1
    
    pdiv = xnode.xpath( '//li/a/text()' )
    
    def pmax(pdiv):
        V = [1]
        for p in pdiv:
            try:
                v = int(p)
                V.append(v)       
            except:
                continue
        return max(V)

    
    return pmax(pdiv)
```

```python
def get_query_page_count(query):
    html = get_html('http://dados.ufrn.br/dataset?q={}'.format(query))
    return get_html_page_count( html )
```

#### Datasets

```python
def get_html_items_href(html):
    xnodes = get_xnodes( html, './/h3', 'class', 'dataset-heading' , '//a/@href' ) 
    return xnodes
```

```python
def get_html_dataset(html):    
    article_html = get_xnode(html , '//article')

    dataset = dict()

    dataset['title'] = strip( get_xnodes( article_html , 'div/h1/text()') ) 
    dataset['key']   = re.sub('[^\w]','_',unidecode(dataset['title']).lower())
    dataset['text']  = strip( get_xnodes(article_html,'//div','class','notes','//p/text()') ) 
    dataset['url']   = get_xnodes( html, '//ol', 'class', 'breadcrumb', '//li/a/@href')

    return dataset
```

```python
def get_html_dataset_license(html):
    xnode = get_xnode( html, '//section','class','license') 

    license_title = xnode.xpath( './/@title' )
    license_href  = xnode.xpath( './/@href')
    license_text  = [text.strip() for text in xnode.xpath( './/text()') if text.strip()]
    license_logo  = ['http://dados.ufrn.br'+p for p in xnode.xpath( './/img//@src')]
    
    license_dict          = dict()
    license_dict['title'] = license_title
    license_dict['href']  = license_href
    license_dict['text']  = license_text
    license_dict['logo']  = license_logo

    return license_dict
```

```python
def get_html_dataset_resources_href(html):
    return get_xnodes(html,'//a','class','heading','/@href')
```

#### Resources

```python
def get_html_resource_info(html):
    tds    = get_xnodes( html, '//table', 'class', 'table-condensed','/tbody//td')
    fields = get_xnodes( html, '//table', 'class', 'table-condensed','/tbody//th/text()')
    values = [strip(get_xnodes(td, './/text()')).strip() for td in tds]

    info_dict        = {f:v for f,v in zip(fields,values)}
    info_dict['url'] = get_xnode( html, '//a','class','resource-url-analytic','/@href')

    return info_dict
```

#### Scrapping

```python
def ufrn_query(entry):
    
    if isinstance(entry, str):
        queries = [entry]
    if isinstance(entry,list):
        queries = entry

    hyperlink = 'http://dados.ufrn.br'

    url = furl( hyperlink )

    resources = []

    for query in queries:
        print('\n',query,end='\n')
        items_href = get_query_items_href( query )

        
        for item_href in items_href:
            url.path = item_href
            print('\n\t', url.url, end='\n\n')
            dataset_html   = get_html( url )
            
            dataset        = get_html_dataset( dataset_html )
            license        = get_html_dataset_license ( dataset_html )
            resources_href = get_html_dataset_resources_href( dataset_html )
            
            resource_url = url.copy()

            for resource_href in resources_href:
                resource_url.path = resource_href
                
                resource_html = get_html( resource_url )  
                resource_dict = get_html_resource_info(resource_html)

                resource_dict['dataset'] = dataset.copy()
                resource_dict['license'] = license

                print('\t\t',resource_dict['url'], end='\n')

                resources.append( resource_dict )
    
    arcevo_dict = dict()
    for resource in resources:
        key = resource['dataset']['key']
        if not key in arcevo_dict: arcevo_dict[key] = []
        arcevo_dict[key].append( resource['url'] )

    
    dict_key_str = lambda key : "arcevo_dict['{}']".format(key)
    n = max([len(dict_key_str(key)) for key in arcevo_dict.keys()])

    arcevo_str = "arcevo_dict = dict()\n\n"
    for key in sorted(arcevo_dict.keys(),key=lambda e: e):
        links = arcevo_dict[key]
        entry = ('{: <' + str(n) + 's} = [' ).format(dict_key_str(key))
        for link in sorted(sorted(links),key=lambda e: os.path.splitext(e)[1],reverse=True):
            entry += "'{link}', ".format(link=link)
        entry += "]\n"
        arcevo_str += entry

    return arcevo, resources, arcevo_str
```
